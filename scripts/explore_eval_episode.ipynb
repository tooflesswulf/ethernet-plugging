```python
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# failed one 
failed_ids = [1, 2, 6, 7, 8, 9, 10]
success_ids = [3, 4, 5]

"""No all failures can be indicated by force,
1) inprecise plugin 
2) bad timeming for unplug
3) ....
* Indicates by image, force, proprioceptions.
* See what happend if I just let it fail?
"""
dir_path = '/home/atkesonlab4/Desktop/YiqiProject/100%_Project/ethernet-plugging/logs_plugin_unplug_50'
def get_force(ep_path):
    loaded = np.load(os.path.join(ep_path, 'states.npz'))
    
    return loaded['force']
```

```python
failed_forces, success_forces = [], []
for failed_id in failed_ids:
    ep_path = os.path.join(dir_path, f'episode{failed_id:06d}')
    failed_forces.append( get_force(ep_path) )

success_forces = []
for success_id in success_ids:
    ep_path = os.path.join(dir_path, f'episode{success_id:06d}')
    success_forces.append( get_force(ep_path) )
```

### Check if force indicate failures accurately?

```python
D = 6
fig, axes = plt.subplots(D, figsize = (12, 10))
for failed_force in failed_forces:
    for d in range(D):
        axes[d].plot(failed_force[:, d], color='red', )
for success_force in success_forces:
    for d in range(D):
        axes[d].plot(success_force[:, d], color='green', )
axes[0].set_title('Forces, Success(green), Failed(red)')
```
